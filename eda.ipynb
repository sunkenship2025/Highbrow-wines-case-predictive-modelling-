# üìä EDA-1 Summary: Hypothesis Testing Results

## üéØ Goal
Validate theoretical framework (6 latent traits) against empirical data patterns.

---

## ‚úÖ Key Findings by Trait

### 1Ô∏è‚É£ Target Distribution (Propensity Problem Confirmed)
**Finding:** Buyers are a **minority class**
- This is a **propensity/ranking problem**, not a balanced classification task
- Modeling approach: Focus on ranking high-propensity customers, not just accuracy

---

### 2Ô∏è‚É£ Digital Comfort (STRONGEST SIGNAL)
**Hypothesis:** Collishop users are more likely to buy highbrow wine online  
**Result:** ‚úÖ **CONFIRMED & AMPLIFIED**

**Key Insight:**  
- Collishop customers show **significantly higher conversion rates**
- Digital trust is the primary barrier, not wine knowledge
- Online grocery shopping = psychological proof of digital comfort

**Implication:** `Collishop_customer` likely to be **#1 or #2 predictor** in model

---

### 3Ô∏è‚É£ Economic Freedom (Directional Effect Confirmed)
**Hypothesis:** Lower price sensitivity ‚Üí higher likelihood  
**Result:** ‚úÖ **CONFIRMED with nuance**

**Key Insight:**  
- Buyers are **on average less price-sensitive** than non-buyers
- But **substantial overlap** exists between groups
- Price sensitivity alone is **insufficient** to distinguish buyers

**Implication:** This is a **secondary signal** or **interaction-based predictor**, not a standalone driver

---

### 4Ô∏è‚É£ Wine/Alcohol Affinity (Channel > Category)
**Hypothesis:** Online alcohol purchases > offline wine purchases  
**Result:** ‚úÖ **STRONGLY CONFIRMED**

**Key Insight:**  
- Online Porto/Whisky purchases are **extremely sparse** but show **very strong separation**
- Buyers are **orders of magnitude more likely** to have prior online premium alcohol purchases
- **Channel behavior (online) predicts better than category behavior (offline wine)**

**Implication:** `cat_AP_STDR_PortoONLINE` and `cat_AP_STDR_WhiskyONLINE` are **gold-standard features** (combines premium + digital + alcohol)

---

### 5Ô∏è‚É£ Culinary Sophistication (VALIDATED)
**Hypothesis:** Gourmet food buyers exhibit higher wine propensity  
**Result:** ‚úÖ **STRONGLY CONFIRMED**

**Key Insight:**  
- Buyers spend **significantly more** in gourmet categories:
  - Tapas
  - Seasonal specialty cheese (KaasSeizoenskazen)
  - Smoked fish (VisGerookt)
- This is a **lifestyle signal**, not just category spend

**Implication:** Premium wine is a **lifestyle complement**. Gourmet food features likely strong predictors.

---

### 6Ô∏è‚É£ Customer Embeddedness (HYPOTHESIS REVISED)
**Initial Hypothesis:** High SOW ‚Üí routine behavior ‚Üí lower exploration  
**Result:** ‚ö†Ô∏è **HYPOTHESIS CORRECTED**

**Key Insight:**  
- High SOW shows **positive relationship** with buying (not negative)
- Interpretation shift:
  - High SOW = deeply embedded in ecosystem
  - High SOW = see Colruyt as "place to buy things"
  - High SOW ‚Üí higher **exposure** ‚Üí higher chance of trial

**For binary target ("did they ever buy?"):**  
High SOW ‚Üí High exposure ‚Üí Higher trial probability

**Implication:** SOW is **positive but may still be non-linear**. Extreme outliers (100%+ SOW) may differ from 60-80% SOW customers.

---

### 7Ô∏è‚É£ Life-Stage Context (Asymmetric Constraints)
**Hypothesis:** Life-stage constraints reduce likelihood  
**Result:** ‚úÖ **PARTIALLY CONFIRMED (asymmetric)**

**Key Insight:**  
- **Young children (Babyluiers):** ‚úÖ Reduces likelihood significantly
  - Active parenting = time/budget constraints
- **Elder care (Incontinentie_luiers):** ‚ùå **No strong effect**
  - Elderly care ‚â† same constraints as young children

**Implication:** Life-stage effects are **heterogeneous**. Some constrain exploration (young families), others do not (elder care). Use as **contextual modifiers**, not primary drivers.
